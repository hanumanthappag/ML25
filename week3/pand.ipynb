### 2) Aggregation & Grouping
Goal: compute summary statistics overall or per group (sum, mean, count, min, max, etc.)

```python
#Aggregation
print(flight['price'].sum()) # sum of the price of all flights

print(flight['duration'].mean()) #Average duration of all flights

print(flight['airline'].count())  # Total no of airlines

print(flight['price'].min()) # Minimum price of flights

print(flight['price'].max()) # Maximum Price

flight.aggregate({'price':['sum','min','max','count','mean'],'duration':['sum','min','max','count','mean']})

#flight.describe()
```

```python
flight.describe()
```

```python
#Grouping

flight.head()
```

```python
grouped=flight.groupby('airline')
groupedbycity=flight.groupby('source_city')
grouped['price'].mean()   # average price of flights of different airlines
```

```python
groupedbycity['price'].mean()  # average price of flights of different source city
```

```python
groupedbyclass=flight.groupby('class')
groupedbyclass['price'].mean()  # average price of flights based on class
```

```python
flight.groupby('airline').agg({'price':['min','max','mean','count'],'duration':['min','max','mean']})
```

```python
flight.groupby('airline').count().sum()
#flight.info()
```

```python
#Group Filtering
flight.groupby('airline').filter(lambda x:x['price'].mean()<5000)
```

```python
cheapflights=flight[flight['price']<5000]
cheapflights.groupby('airline').count()
```

### 3) Time series operations

Create time index example

```python
dates = pd.date_range('2025-01-01', periods=8, freq='D')  # daily
#ts = pd.DataFrame({'Sales': [200,220,210,230,250,240,200,300]}, index=dates)
ds=pd.DataFrame({'Sales':[200,220,300,400,500,600,200,300]},index=dates)
print(ds)
```

###Resample  change frequency (aggregation by period)

```python
# Sum per week (default week end is Sun)
weekly = ds.resample('W').sum()
print(weekly)
# or monthly: ts.resample('M').sum()
```

### Rolling (moving) window

```python
# 3-day moving average
ds['MA_3'] = ds['Sales'].rolling(window=3).mean()
# First two rows will be NaN because window is 3
ds
```

###Shift / lag  to compute day-over-day changes

```python
ds['PrevDay'] = ds['Sales'].shift(1)            # previous day's sales
ds['Pct_change'] = ds['Sales'].pct_change()    # percent change from prev row
ds
```

### 4) Pivot & Melt (reshaping)

```python
df = pd.DataFrame({
    'Date': ['2025-01-01','2025-01-01','2025-01-02','2025-01-02'],
    'City': ['A','B','A','B'],
    'Sales': [100,150,200,250]
})
print(df)
pivot = df.pivot(index='Date', columns='City', values='Sales')
print("\n\n")
print(pivot)
# pivot:
# City         A    B
# Date
# 2025-01-01 100  150
# 2025-01-02 200  250
```

```python
pivot2 = df.pivot_table(index='Date', columns='City', values='Sales', aggfunc='sum', margins=True)
# margins=True adds totals
pivot2
```

```python
m = pivot.reset_index().melt(id_vars='Date', var_name='City', value_name='Sales')
m
# converts pivot back to original long form
```

```python
#Example 1  Sales by Month and Product
df = pd.DataFrame({
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar'],
    'Product': ['Laptop', 'Mobile', 'Laptop', 'Mobile', 'Laptop', 'Mobile'],
    'Revenue': [100000, 80000, 120000, 85000, 90000, 95000]
})  # 5*3 Shape
print(df)
pivot_df = df.pivot(index='Month', columns='Product', values='Revenue') # 3*5 shape
print("\n\n",pivot_df)  
```

```python
m1 = pivot_df.reset_index().melt(id_vars='Month', var_name='Product', value_name='Revenue')

print(m1)
```

### 5) Map, Filter, Reduce, Lambda with Pandas DataFrames

```python
df = pd.DataFrame({'Product':['A','B','C'],'Category':['X','X','Y']})
print(df)
df['CategoryCode'] = df['Category'].map({'X':1,'Y':2})
print(df)
# or map a function
df['Upper'] = df['Product'].map(lambda s: s.lower())
print(df)
```

```python
# filter
filtered = df[df['CategoryCode'] > 1]

print(filtered)
# or df.query('Sales > 150') for readable string expression
```

### reduce  from functools

```python
from functools import reduce

df = pd.DataFrame({
    'Date': ['2025-01-01','2025-01-01','2025-01-02','2025-01-02'],
    'City': ['A','B','A','B'],
    'Sales': [100,150,200,250]
})

total = reduce(lambda a, b: a + b, df['Sales'])
print(total)
# equivalent to df['Sales'].sum()  prefer sum()
```

```python
#Lambda functions

df['Discounted'] = df['Sales'].map(lambda x: x * 0.9)
print(df)
```
